```python
#renaming columns to shorter names for ease of keeping track
data = data.rename(columns={'2. Your Age':'Age',
                           '3. Gender': 'Gender',
                           '4. What is your monthly income?':'Monthly Income',
                           '5. Which fashion accessory do you most frequently buy? (select multiple if needed)':'Frequently bought',
                           '6. What is your most preferred medium when you are shopping for fashion accessories? ':'Preferred medium',
                            '7. How much do you usually spend on shopping (at one go) for fashion accessories?':'Amount spent on shopping',
                           '8. Do you prefer shopping alone or with company?':'Alone or with Company',
                           '9. Are you a fan of window shopping?':'Likes window shopping',
                           '10. Around which area do you live?':'location',
                           '11. Do you shop for fashion accessories nearby your house or look elsewhere?':'Shop near house or elsewhere',
                           '12. Which is your most preferred shopping mall to go to when you go shopping for fashion accessories?':'Preferred shopping mall',
                           '13. Which of the following Offline Shops have you most often visited? (select multiple if needed)':'Most visited offline shops',
                           '14. Why are you a fan of the preferred malls you have selected? (select multiple if needed)':'Reason for preferred malls',
                           '15. Rate the ambience of your most preferred shopping mall. ':'Preferred mall ambience rating',
                           '16. How would you rate the convenience for transportation availability of your most preferred shopping mall?':'Convenience rating pref mall',
                           '17. Have you ever gone into a shopping mall and came out without shopping for anything?':'Visited mall without shopping',
                           '18. What is your most acceptable time to go shopping?':'Preferred shopping time',
                           '19. Would you prefer a 24/7 shopping mall?':'24/7 shopping mall preference',
                           '20. Are you a brand specific person when shopping for fashion accessories?':'Specific brand preference',
                           '21. Do you feel annoyed when you cannot find the store of your desired brand in a shopping mall?':'Annoyance when desired brand not available',})
data
```

```python
#location, timestamp and reason for preferred malls seem irrelevant to predict our target variable
data1=data.drop(columns=['location'])
data1=data1.drop(columns=['Timestamp'])
data1=data1.drop(columns=['Reason for preferred malls'])
data1
```

```python
data1['Preferred medium']=data1['Preferred medium'].replace('Offline Shopss', 'Offline Shops')
```

```python
#knowing data types is important to keep track of categorical vs numerical variables
data.dtypes
```

```python
#original data set contained comma seperated values under most visited offline shops.
#some inputs had different spelling for the same mall
#this overexaggerated the number of unique values. To keep things simple, only the first choice of the respondents has been assumed to be the representative answer
#Hence, we have seperated all the values on ',' and only kept the first value
data1['Most visited offline shops'] = data1['Most visited offline shops'].str.split(',').[str0]
data1['Frequently bought'] = data1['Frequently bought'].str.split(',').str[0]
data1
```

```python
#Now that we have our data all cleaned up, we import the tools that will help us predict preferred medium for shopping.
# All the tools for our predictive model comes from a collection of library called sklearn. There are many models that we can use. But since the thing we want to predict is not number, a classification technique is suitable
#Random forest classifier is most suitable here because the factors for predicting our target are mostly non-numeric.
#Other models like regression require those factors to be tested on various grounds before putting them onto the model. And because of the abundance of  non-numeric variables in our data, we cannot test them without delving into complicated process
#Random Forest eliminated this problem because it can work with non-linear data and is more powerful when it comes to dataset like this
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import mean_squared_error, r2_score
```

```python
#Here we have seperated variables our dataset have based on their types. This is important because our model cannot understand both types of variables. So we need to convert them in a machine readable way.
categorical_features = data1.select_dtypes(include=['object']).columns.tolist()
numeric_features = data1.select_dtypes(exclude=['object']).columns.tolist()
```

```python
#isolating the target variable from the rest of the independent variables because we don't want it to undergo any changes since this is the thing we want to predict.
if 'Preferred medium' in categorical_features:
    categorical_features.remove('Preferred medium')
```

```python
#Now we keep the target variable in a safe place called y
y = data1['Preferred medium'].copy()
data1_features = data1.drop('Preferred medium', axis=1)
```

```python
#transforming each value of the categorical features (meaning the independent variables that have an effect on our target) into seperate columns with binary values so that the model can interpret it better
preprocessor = ColumnTransformer(
    transformers=[
        ('num', SimpleImputer(strategy='mean'), numeric_features),
        ('cat', OneHotEncoder(handle_unknown='ignore'), categorical_features)
    ])
```

```python
#before we go with the prediction we need to empower our model with several precautionary tools. This is packed into a bag called model_pipleline.
#Preprocessor is one tool in the bag that contains our 'transformed' data. By 'transformed', we mean the changes our data underwent in order to be machine-readable.
#Scaler ensures that all our data are on the same page. Since our data contains both binary and decimal numbers, our machine could get confused if we didn't bring all the data in the same scale
model_pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('scaler', StandardScaler(with_mean=False)),
    ('model', RandomForestClassifier(random_state=42,))
])
```

```python
#splitting data into train and test groups. Training data will be fed into the model so it can learn the pattern. And test data is reserved to compare how well our model is predicting
#70-30 split is chosen. So 70% of the data will be used for training, rest for testing. A random state is chosen for result reproducibility
X_train, X_test, y_train, y_test = train_test_split(data1_features, y, test_size=0.3, random_state=123)
model_pipeline.fit(X_train, y_train)
```

```python
#importing the modules required for performance evaluation. So these tools can help us understand how well our model is at predicting
from sklearn.metrics import accuracy_score, precision_score, recall_score, confusion_matrix
```

```python
#defining y_pred as the predicated values after the model has been trained against test data
y_pred = model_pipeline.predict(X_test)
```

```python
# Compute accuracy
accuracy = accuracy_score(y_test, y_pred)

# Compute precision
precision = precision_score(y_test, y_pred, average='micro')

# Compute recall
recall = recall_score(y_test, y_pred, average='micro')


# Compute confusion matrix
conf_matrix = confusion_matrix(y_test, y_pred)
```

```python
#Now we are going to get back the evaluation scores from the code
#We can see that accuracy, precision and recall are 75.4%
#This means that our model is correct at predicting 'preferred medium' variable 75 times out of 100.
#Furthermore, suppose the model says 100 people like online shopping and 100 like offline shopping. In that case, this model will have picked correctly predicted the preference of 75% for both offline and online preference. That's what precision means
#Finally, recall=75.4% means that if in reality, only 100 people liked offline shops, our model will correctly identify 75 of them.
print(accuracy)
print(precision)
print(recall)
print(conf_matrix)
```

```python
#We fed a lot of 'prcoessed' information into our model. Now we will extract them out to see which ones were very important to predict people's preferred shopping modality
feature_names = np.concatenate([
    preprocessor.named_transformers_['num'].get_feature_names_out(numeric_features),
    preprocessor.named_transformers_['cat'].get_feature_names_out()
])
```

```python
#These are the codes to get those 'important information'
importances = model_pipeline.named_steps['model'].feature_importances_


feature_importance_df = pd.DataFrame({'Feature': feature_names, 'Importance': importances})


top_features = feature_importance_df.sort_values(by='Importance', ascending=False).head(30)
```
